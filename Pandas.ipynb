```python
colours = pd.Series(
    [
        "Red" , "Blue" ,  "White"
    ]
)
```

## Series = 1-dimensional (column)

## DataFrame = 2-dimensional (rows and columns)

## DataFrames can take into series.

```python
# DataFrames take into python dictionaries.
car_data = pd.DataFrame(
    {
        "Car make": series ,
        "Colour": colours
    }
)
car_data
```
