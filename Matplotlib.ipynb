## Integrates directly with pandas

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot(); # This is our figure
```

```python
x = [ 1 , 2 , 3 , 4]
y = [ 11 , 22 , 33 , 44 ]
plt.plot( x , y );
```

```python
# First method
fig = plt.figure() # creates a figure
ax = fig.add_subplot() #adds some axes
plt.show()
```

```python
# Second method
fig = plt.figure()
ax = fig.add_axes([ 1 , 1 , 1 , 1])
ax.plot(x , y) # Add some data
plt.show()
```

```python
# Third method(recommended)
fig , ax = plt.subplots()  # Subplot aynı anda farklı grafikleri yanyana cizdirmek.
ax.plot(x , y);
```

## Matplotlib example workflow
1. Import matplotlib.pyplot
2. Prepare data ( x = [ 1 , 2 , 3] , y = [ 50 , 45 , 72] )
3. Setup plot

**fig , ax = plt.subplots(figsize = (10 , 10))  , figsize = (height , width)**
4. Plot data

**ax.plot( x , y )**
3 te figure yani bütün grafiği kapsayan bölgeyi oluşturup axisleri oluşturduk şimdi axislerimize datalarımızı yerleştiriyoruz.

5. Customize plot

**ax.set(title = "Simple Plot" ,**
**xlabel = "x-axis" ,**
**ylabel = "y-axis")**


## Making figure with numpy arrays
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
# Create some data
x = np.linspace( 0 , 10 , 100 )
x
```

```python
# Plot the data with line plot  ( plot() default is line )
fig , ax = plt.subplots()
ax.plot( x , x**2 );
```

```python
# Use same data for scatter plot
fig , ax = plt.subplots(figsize = (10 , 7 ))
ax.scatter( x , x**2 );
```

```python
# Another scatter plot
fig , ax = plt.subplots()
ax.scatter(x , np.sin(x));
```

```python
# Make a plot from dictionaries
nut_butter = {
    "Almond" : 10 ,
    "Peanut": 8 ,
    "Cashew": 12
}
fig , ax = plt.subplots()
ax.bar(nut_butter.keys() , nut_butter.values() )
ax.set(
    title = "butter prices" ,
    xlabel = "Butters" ,
    ylabel = "Price"
);
```

```python
# For horizontal bar plot
fig , ax = plt.subplots()
ax.barh( list( nut_butter.keys() ) , list( nut_butter.values() ) );  # barh horizontal plot için. barh parametreleri liste şeklinde alıyor bu yüzden list() methodu ile listeye ceviriyoruz.
```

```python
# Histogram
# Make some data
x = np.random.randn(1000)  # Normal distribution randn
fig , ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# First option for subplot
fig , ( ( ax1 , ax2 ) , ( ax3 , ax4 ) ) = plt.subplots(
    nrows = 2 ,
    ncols = 2 ,
    figsize = ( 10 , 5 )
)
# Plot to each different axis
ax1.plot( x , x/2 );
ax2.scatter(np.random.random(10) , np.random.random(10) );
ax3.bar(nut_butter.keys() , nut_butter.values());
ax4.hist(np.random.randn(1000));
```
